```python
image_path = '/content/image/beatles01.jpg'
```

##### **1) PIL(Python Imaging Library)**

```python
import matplotlib.pyplot as plt
from PIL import Image

image_pil = Image.open(image_path)
print(f'image_pil type : {type(image_pil)}, {image_pil.size}')

plt.figure(figsize=(6, 4))
plt.imshow(image_pil)
plt.show()
```

##### **2) scikit-image**

```python
from skimage import io

image_skimage = io.imread(image_path)
print(f'image_skimage type : {type(image_skimage)}, {image_skimage.shape}')

plt.figure(figsize=(6, 4))
plt.imshow(image_skimage)
plt.show()
```

##### **3) OpenCV**

```python
import cv2

image_opcv = cv2.imread(image_path)
image_opcv_coverted_01 = cv2.cvtColor(image_opcv, cv2.COLOR_BGR2RGB)
cv2.imwrite('/content/image/beatles02.jpg', image_opcv)
image_opcv_coverted_02 = plt.imread('/content/image/beatles02.jpg')

fig, axs = plt.subplots(figsize=(18, 6), nrows=1, ncols=3)
axs[0].imshow(image_opcv)
axs[1].imshow(image_opcv_coverted_01)
axs[2].imshow(image_opcv_coverted_02)

axs[0].set_title('imread(BGR)')
axs[1].set_title('cvtColor(BGR > RGB)')
axs[2].set_title('imwrite(RGB)')
plt.show()
```
