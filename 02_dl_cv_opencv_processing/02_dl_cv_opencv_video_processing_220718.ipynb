### **OpenCV : video processing**

```python
import cv2

video_input_path = '/content/Night_Day_Chase.mp4'
video_output_path = '/content/drive/MyDrive/Colab Notebooks/update/dl_cv/2_dl_cv_opencv_processing/data/Night_Day_Chase_output.mp4'

cap = cv2.VideoCapture(video_input_path)
codec = cv2.VideoWriter_fourcc(*'XVID')
video_fps = round(cap.get(cv2.CAP_PROP_FPS))
video_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)), round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

video_writer = cv2.VideoWriter(video_output_path, codec, video_fps, video_size)
print(f'frame size : {frame_count}, video fps : {video_fps}, video size : {video_size}')
```

```python
rgb_green = (0, 255, 0) 
rgb_red = (0, 0, 255)

index = 0
while True:
  hasframe, frame = cap.read()
  if not hasframe:
    print('process completed')
    break
  else:
    index += 1
    print(f'frame : {index}')

    caption = f'frame : {index}'
    cv2.rectangle(frame, (300, 100, 800, 400), color=rgb_green, thickness=4)
    cv2.putText(frame, caption, (300, 95), cv2.FONT_HERSHEY_COMPLEX, 0.7, color=rgb_red, thickness=1)
    video_writer.write(frame)

video_writer.release()
cap.release()
```
