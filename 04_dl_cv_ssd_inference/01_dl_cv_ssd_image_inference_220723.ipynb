```python
draw_image = image.copy()

rgb_green = (0, 255, 0)
rgb_red = (0, 0, 255)

for detection in cv_out[0, 0, :, :]:
  class_id = int(detection[1])
  class_confidence = detection[2]
  if class_confidence > 0.5:
    left = detection[3] * image_width
    top = detection[4] * image_height
    right = detection[5] * image_width
    bottom = detection[6] * image_height

    caption = '{} : {:.4f}'.format(labels_to_names[class_id], class_confidence)
    cv2.rectangle(draw_image, (int(left), int(top)), (int(right), int(bottom)), color=rgb_green, thickness=2)
    cv2.putText(draw_image, caption, (int(left), (int(top)-5)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, color=rgb_red, thickness=1)

draw_image = cv2.cvtColor(draw_image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12, 10))
plt.imshow(draw_image)
plt.show()
```

```python
import time

def get_detected_image(cv_net, image, threshold, use_copied_image=True, print_time=True):
  start = time.time()

  image_height = image.shape[0]
  image_width = image.shape[1]
  print(f'image shape : {image.shape}')
  print(f'image height : {image_height}, image width : {image_width}')

  draw_image = None
  if use_copied_image:
    draw_image = image.copy()
  else:
    draw_image = image

  cv_net.setInput(cv2.dnn.blobFromImage(image, swapRB=True, crop=False))
  cv_out = cv_net.forward()
  print(f'cv_out shape : {cv_out.shape}')

  rgb_green = (0, 255, 0)
  rgb_red = (0, 0, 255)

  for detection in cv_out[0, 0, :, :]:
    class_id = int(detection[1])
    class_confidence = float(detection[2])
    if class_confidence > threshold:
      left = detection[3] * image_width
      top = detection[4] * image_height
      right = detection[5] * image_width
      bottom = detection[6] * image_height

      caption = '{} : {:.4f}'.format(labels_to_names[class_id], class_confidence)
      cv2.rectangle(draw_image, (int(left), int(top)), (int(right), int(bottom)), color=rgb_green, thickness=2)
      cv2.putText(draw_image, caption, (int(left), (int(top)-5)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, color=rgb_red, thickness=1)

  if print_time:
        print(f'detection time : {round(time.time()-start, 2)}초')

  return draw_image
```

```python
image = cv2.imread('./image/beatles01.jpg')
draw_image = get_detected_image(cv_net, image, threshold=0.5, use_copied_image=True, print_time=True)
data_dir = ('./drive/MyDrive/Colab Notebooks/update/dl_cv/4_dl_cv_ssd_inference/data/beatles01_output.jpg')
cv2.imwrite(data_dir, draw_image)
```

```python
plt.figure(figsize=(12, 10))
image_output = plt.imread(data_dir)
plt.imshow(image_output)
plt.show()
```
