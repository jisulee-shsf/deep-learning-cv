```python
import cv2
import matplotlib.pyplot as plt

image = cv2.cvtColor(cv2.imread('/content/data/audrey01.jpg'), cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 8))
plt.imshow(image)
plt.show()
```

```python
import selectivesearch

_, regions = selectivesearch.selective_search(image, scale=100, min_size=3000)
print(f'{type(regions)} / {regions[0]}')
```

```python
regions_rects_01 = [region['rect'] for region in regions]
print(len(regions_rects_01), regions_rects_01[0])
```

```python
rgb_green = (125, 255, 51)
image_copy = image.copy()
for i in regions_rects_01:
  left = i[0]
  top = i[1]
  right = left + i[2]
  bottom = top + i[3]
  image_copy = cv2.rectangle(image_copy, (left, top), (right, bottom), color=rgb_green, thickness=2)

plt.figure(figsize=(10, 8))
plt.imshow(image_copy)
plt.show()
```

```python
regions_rects_02 = [region['rect'] for region in regions if region['size'] > 20000]
print(len(regions_rects_02), regions_rects_02[0])
```

```python
rgb_green = (125, 255, 51)
image_copy = image.copy()
for i in regions_rects_02:
  left = i[0]
  top = i[1]
  right = left + i[2]
  bottom = top + i[3]
  image_copy = cv2.rectangle(image_copy, (left, top), (right, bottom), color=rgb_green, thickness=2)

plt.figure(figsize=(10, 8))
plt.imshow(image_copy)
plt.show()
```
